# EDA

Lo primero que se hizo fue descargar los datos del link entregado al inicio del proyecto, luego de eso se procedió a preparar los datasets en el EDA, en primera instancia se analizó el dataset correspondiente a la conectividad, al encontrarse que habían datos duplicados se procedió a eliminar aquellas filas duplicadas que no tenían valor y a juntar los datos de dos filas cuando se lo creyó conveniente, luego se comenzó a cambiar los tipos de datos en las columnas que se creyó necesario, también se eliminaron las filas con datos de población nulos porque afectarían de mala manera al realizamiento de los KPI's.

Después se procedió con el dataset correspondiente a la penetración, al cual se le cambió los tipos de datos en las columnas que se creyó necesario.

Luego se procedió con el dataset correspondiente al historico_velocidad_internet, al cual también se le cambió los tipos de datos en las columnas que se creyó necesario y se le cambió el nombre de la columna Trimestre.1 por Periodo.

El siguiente fué el dataset correspondiente a ingresos, al cual lo único que se le hizo fué cambiarle los tipos de datos en las columnas que se creyó necesario.

El próximo fué el dataset correspondiente a accesos_a_internet, el cuál tuvo el mismo procedimiento que el anterior.

El último dataset fué el correspondiente a baf, al que se lo trato de la misma manera que los anteriores.
