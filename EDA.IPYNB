# 5 dataset AccesosaInternetfijoporvelocidaddebajadaylocalidad_2776171689464891656.csv

Se instancia el dataset como un dataframe en la variable accesos_a_internet

```python
accesos_a_internet = pd.read_csv('Datasets/AccesosaInternetfijoporvelocidaddebajadaylocalidad_2776171689464891656.csv')
accesos_a_internet
```

```python
accesos_a_internet.info()
```

Se analiza si hay datos repetidos

```python
duplicados = accesos_a_internet.duplicated(subset=['Provincia', 'Partido', 'Localidad'], keep=False)
filas_duplicadas = accesos_a_internet[duplicados]
filas_duplicadas
```

CAMBIAR LAS COLUMNAS DE MBPS QUE SON OBJECT A FLOAT

Se crea una variable que contenga todas las columnas mbps

```python
columnas_convertir = [col for col in accesos_a_internet.columns if col not in ['Provincia', 'Partido', 'Localidad', 'Link Indec', 'Otros']]
```

Para cambiar el tipo de columna, se itera por cada una de ellas, con el condicionante if tomamos solamente las que son de tipo object, una vez hecho eso se les cambia la ',' que aparece un algunas filas por un '.' para que sea posible la transformación a float y por último se realiza dicha transformación

```python
for columna in columnas_convertir:
    if accesos_a_internet[columna].dtype == object:
        accesos_a_internet[columna] = accesos_a_internet[columna].str.replace(',', '.')
        accesos_a_internet[columna] = pd.to_numeric(accesos_a_internet[columna])
```

```python
accesos_a_internet.info()
```
